```
from abc import ABC, abstractmethod

class AbstractEmployee(ABC):
    new_id = 1

    @abstractmethod
    def say_id(self):
        pass


class Employee(AbstractEmployee):
    def __init__(self, company_name):

        self.id = AbstractEmployee.new_id

        AbstractEmployee.new_id += 1

        self._company = company_name

        self._name = None

    def say_id(self):

        print(f"My id is {self.id}")


    def get_name(self):
        return self._name


    def set_name(self, name):
        self._name = name


    def del_name(self):
        del self._name


class User:
    def __init__(self, username, role):
        self.username = username
        self.role = role

    def say_user_info(self):
        print(f"Username: {self.username}, Role: {self.role}")


class Admin(Employee, User):
    def __init__(self):

        Employee.__init__(self, "MyCompany")

        User.__init__(self, f"admin{self.id}", "Admin")

    def say_id(self):

        print("I am an Admin")

        Employee.say_id(self)


class Manager(Admin):
    def say_id(self):

        print("I am in charge")

        Admin.say_id(self)


e1 = Employee("MyCompany")
e2 = Employee("MyCompany")
e3 = Admin()
e4 = Manager()


e1.say_id()
e2.say_id()
e3.say_id()
e4.say_id()


e3.say_user_info()

meeting = [e1, e3, e4]

for attendee in meeting:
    attendee.say_id()

class Meeting:
    def __init__(self):
        self.attendees = []

    def __add__(self, employee):

        self.attendees.append(employee)
        print(f"Employee with ID {employee.id} added successfully.")

    def __len__(self):

        return len(self.attendees)


m1 = Meeting()
m1 + e1
m1 + e2
m1 + e3

print(len(m1))

#print(dir(e1))
```
